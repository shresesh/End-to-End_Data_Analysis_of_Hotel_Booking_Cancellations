Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

Loading the dataset

```python
df = pd.read_csv('hotel_bookings 2.csv')
```

```python
# Parse reservation_status_date considering day-first format (e.g., '22/4/2015').
# Use dayfirst=True and infer formats; invalid parses will become NaT so we can inspect them.
df['reservation_status_date'] = pd.to_datetime(
	df['reservation_status_date'],
	dayfirst=True,
	infer_datetime_format=True,
	errors='coerce'
)

# Fallback: try parsing remaining unparsed values assuming month-first if dataset has mixed formats
mask = df['reservation_status_date'].isna()
if mask.any():
	df.loc[mask, 'reservation_status_date'] = pd.to_datetime(
		df.loc[mask, 'reservation_status_date'],
		dayfirst=False,
		infer_datetime_format=True,
		errors='coerce'
	)

# Report any remaining unparsed values
remaining = df['reservation_status_date'].isna().sum()
if remaining:
	print(f"Warning: {remaining} reservation_status_date values could not be parsed. Examples:")
	print(df.loc[df['reservation_status_date'].isna(), 'reservation_status_date'].unique()[:10])
else:
	print("All reservation_status_date values parsed successfully.")
```

```python
# Drop columns if they exist to avoid KeyError when columns are missing
df.drop(['company', 'agent'], axis=1, inplace=True, errors='ignore')
df.dropna(inplace=True)
```

```python
df = df[df['adr']<5000]
```

Data Anaysis and Visualization

Reservation Status Count

```python
cancelled_precent = df['is_canceled'].value_counts(normalize=True)
print(cancelled_precent)

plt.figure(figsize=(5,4))
plt.title('Reservation stauts count')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts(), edgecolor='k', width =0.7)
plt.show()
```

Reservation Status Count for both the hotels

```python
plt.figure(figsize=(8,4))
axl = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status by hotel type', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not canceled', 'canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

Average Daily Rate in both the hotels

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
plt.figure(figsize=(12,6))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=16)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=14)
plt.show()
```

Monthy Reservation Status

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(10,5))
axl = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels, _ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1, 1))
plt.title('Monthly reservation status', size = 20)  
plt.xlabel('Month')
plt.ylabel('number of reservations')
plt.legend(['Not canceled', 'canceled'])
plt.show()
```

Average Daily Rate for every month

```python
plt.figure(figsize=(10,6))
plt.title('ADR per month', fontsize = 16)
# Aggregate canceled reservations by month, then pass explicit x/y/data to avoid argument ambiguity
agg = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month', y='adr', data=agg)
plt.show()

```

Countries with most cancellations

```python
cancelled_data = df[df['is_canceled']==1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(10,6))
plt.title('Top 10 countries by canceled bookings', fontsize=16)
plt.pie(top_10_countries, autopct = '%.2f', labels=top_10_countries.index)
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values(by='reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values(by='reservation_status_date', inplace=True)

plt.figure(figsize=(12,6))
plt.title('Average Daily Rate') 
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Canceled Reservations')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled Reservations')
plt.legend(fontsize=14)
```
